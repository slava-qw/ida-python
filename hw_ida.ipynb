````python
N = 2
A = [0, 1, 2, 3, 4, 5]



def search(n, a):
  """
  Find number n in array a using binary search with recursion (O(log(len(a))))
  """
  
  if len(a) == 1:
      if n == a[0]:
          return a[0]
      else:
        return
  if n <= a[len(a)//2 - 1]:
    a = a[:len(a)//2]
  else:
    a = a[len(a)//2:]
  return search(n, a)

# some problems with N, when N - A[i] = N / 2
# if N / 2 % 10 == 0:
# else:


# Find number N - A[i], if such number exists, add it to the l together with the A[i]
l = []
for i in range(len(A)):
    num = N - A[i]
    if num >= N//2:
      scn_num = search(num, list(A[:i] + A[i + 1:]))
      if scn_num != None:
        l.append((A[i], scn_num))
    else:
        break

if bool(l):
  print(f'There are some numbers that fit the condition: {l}')
else:
  print('There are no such numbers')
````

## 3.3

**Три числа с суммой $N$**

Пусть дан отсортированный список чисел `[10, 10, 15...]` и некоторое число $N$. Проверить, есть ли в нем $3$ числа, которые в сумме дают $N$. Вывести такие "*триплеты*" чисел.

````python
def search2_pt(n, a):
  """
  Find number n in array a using the two-pointer technique (O(len(a)))
  """
  l = []
  i = 0
  j = len(a) - 1
  while len(a[i:j+1]) != 1:
    if a[i] + a[j] > n:
      j -= 1
    elif a[i] + a[j] < n:
      i += 1 
    else:
      l.append((a[i], a[j]))
      if len(a[i:j + 1]) == 2:
          break
      else:
          i += 1
          j -= 1
  return l


N = 4
A = [0, 0, 0, 1, 1, 1, 2, 2, 3, 4, 5, 7, 8, 9, 10]
s = []

for i in range(len(A)):
    num = N - A[i]
    B = A[:i] + A[i + 1:]
    if num > N//2:
        s.append((A[i], search2_pt(num, B)))
    else:
        break

# make the found answer looks better
p = []
for i in range(len(s)):
  for j in range(len(s[i][1])):
    p.append((s[i][0], s[i][1][j][0], s[i][1][j][1]))
if bool(p):
  print(f'There are some triplets that fit the condition: {set(p)}') # проблема удаления одинаковых триплетов, которые есть в списке
else:
  print('There are no such triplets')
````

````python
# код для уникальных триплетов
# (такой же как и выше только для поиска двух чисел используется линейный алгоритм и в конце удаляются повторяющиеся элементы)
def search2(n, a):
  """
  Find all pairs of numbers in array a, which sum is equal n using the modified two-pointer technique (O(len(a)))
  """
  l = []
  i = 0
  j = len(a) - 1
  while len(a[i:j+1]) != 1:
    if a[i] + a[j] > n:
      j -= 1
    elif a[i] + a[j] < n:
      i += 1 
    else:
      l.append((a[i], a[j]))
      if len(a[i:j + 1]) == 2:
          break
      else:
          i += 1
          j -= 1
  return l


N = 4
A = [0, 0, 1, 1, 1, 2, 2, 3, 4, 5, 7, 8, 9, 10]
s = []

# Find number num = N - A[i], which can be present as sum of two numbers from array A, if so, add them to the s together with the A[i]
for i in range(len(A)):
    num = N - A[i]
    B = A[:i] + A[i + 1:]
    if num > N//2:
        s.append((A[i], search2(num, B)))
    else:
        break

if bool(s):
  # make the found answer looks better
  p = []
  for i in range(len(s)):
    for j in range(len(s[i][1])):
      p.append((s[i][0], s[i][1][j][0], s[i][1][j][1]))

  # find the unique triplets
  t = sorted(list(map(sorted, set(p))))
  i = 0
  while i != len(t) - 1:
    if t[i] == t[i + 1]:
      del t[i]
    else:
      i += 1
  print(f'There are some unique triplets that fit the condition: {t}')
else:
  print('There are no such triplets')
````

````python
from math import log
def preprocess_text(text: str):
  word = r"(\w[\w']*\w|\w)" # also fits for words with 's
  return re.findall(word, text)


def preprocess_corpus(corpus: str):
  '''
  Split corpus of text to sentences.
  '''
  return re.split(r'(?<=\w[:;.!?]) ', corpus) 
  # проблемы с разделением вот такого предложения:
  # 'Как выжить в современной России. У нас есть три способа: 1) стать жертвой маньяка; 2) стать жертвой другого маньяка, которого мы убьем; 3) стать жертвой не маньяка.',
  

def calc_tf(term: str, text: str):
  '''
  Claculate tf for term in the sentence.
  '''
  sentences = preprocess_corpus(text)
  i = 0
  sentence = ''
  for s in sentences:
    if s.find(term) != -1:
      sentence = sentences[i]
      break
  words = preprocess_text(sentence)
  return 0.5 + 0.5 * (words.count(term) / len(words))


def calc_idfs(corpus):
  whole_text = ' '.join(corpus)
  A = preprocess_corpus(whole_text)
  N = len(A)
  k = 0
  d = defaultdict(int)
  
  for sent in A:
    for word in preprocess_text(sent):
      if word.isalpha():
        d[word.lower()] += 1

  for word in d:
    c = 0
    for sent in A:
      if word in sent:
        c += 1

    d[word] = log(N / (1 + c))

  return d

def calc_tfidf(term: str, text: str, precomputed_idfs: dict):
    return calc_tf(term, text) * precomputed_idfs[term]
````

````python
from functools import lru_cache
import time
````

````python
def clock(active=True, unit='s'):

    def decorate(func):
        
        if unit == 's':
            modifier = 1
        elif unit == 'm':
            modifier = 60
        elif unit == 'ms':
            modifier = 0.001
        elif unit == 'h':
            modifier = 3600
        
        
        @wraps(func)
        def clocked(*args, **kwargs):
            if not active:
                return func(*args, **kwargs)
            
            t_start = time.time()
            res = func(*args, **kwargs)
            total_time = time.time() - t_start

            print(f'{func.__name__}({args}) -> {res} executed in {total_time / modifier:.3f} {unit}')

            return res

        return clocked
    
    return decorate
````

````python
from functools import wraps

def my_cache_deco(N=None, K=None, cache_type=1):

  def decororate(func):
    cache = {}
    root = []
    root[:] = [root, root, None, None]
    PREV, NEXT, KEY, RESULT = 0, 1, 2, 3
    full = False


    @wraps(func)
    def mem(*args):
      if N == None: # no cache
          result = func(*args)
          return result

      elif N == -1:  # cache without limits
          key = args[0]
          result = cache.get(key, None)
          if result != None:
              return result
          
          result = func(*args)
          cache[key] = result
          return result

      
      if cache_type == 1:
          nonlocal root, full
          key = args[0]
          link = cache.get(key, None)

          if link is not None:
              link_prev, link_next, _key, result = link
              link_prev[NEXT] = link_next
              link_next[PREV] = link_prev
              last = root[PREV]
              last[NEXT] = root[PREV] = link
              link[PREV] = last
              link[NEXT] = root
              return result
          
          result = func(*args)

          if key in cache:
              pass
          elif full:
              oldroot = root
              oldroot[KEY] = key
              oldroot[RESULT] = result
              root = oldroot[NEXT]
              oldkey = root[KEY]
              oldresult = root[RESULT]
              root[KEY] = root[RESULT] = None
              del cache[oldkey]
              cache[key] = oldroot
          else:
              last = root[PREV]
              link = [last, root, key, result]
              last[NEXT] = root[PREV] = cache[key] = link
              full = (len(cache) >= N)
          return result

      if cache_type == 2:
          pass
        

    return mem
  
  return decororate
````

````python
@clock()
@my_cache_deco(N=4, cache_type=1)
def strange_factorial(n):
  time.sleep(1)
  if n == 0:
    return 1
  elif n == 1:
    return 1
  elif n == 2:
    return 2
  
  return strange_factorial(n - 1) + strange_factorial(n - 1)
````

Сгенерируйте случайный пароль длины N. Убедитесь, что получившийся пароль на самом деле хороший. В нашем случае это:

* Содержит хотя бы две заглавные буквы подряд
* Содержит хотя бы две строчные буквы подряд
* Не содержит пробельные спецсимволы (\t, \n и подобные)
* Не содержит трех цифр подряд
* Содержит и русские, и английские буквы

Ваш код должен генерировать пароли, пока они не удовлетворят заданным условиям.

hint: используйте модуль random

````python
import string
import random
import re


cyrillic_lower = [(lambda c: chr(c))(i) for i in range(1072, 1104)]
cyrillic_upper = [(lambda c: chr(c))(i) for i in range(1040, 1072)]
cyrillic_ansi = cyrillic_lower + cyrillic_upper
cyrillic_ansi_str = ''.join(cyrillic_ansi)

allowed_symbols = string.ascii_letters + string.digits + cyrillic_ansi_str


def password_generation(N=10):
    password = "".join([random.choice(allowed_symbols) for n in range(N)])
    return password


def check_password(password):
    pattern1 = re.compile('([A-Z]{2,}|[А-Я]{2,})')  # test1: True
    pattern2 = re.compile('([a-z]|[а-я]){2,}')  # test2: True
    pattern_zero = re.compile('\d{3,}')  # false
    pattern3 = re.compile('[A-Za-z]')  # true
    pattern4 = re.compile('[А-Яа-я]')  # true
    tests = {
        0: pattern_zero,
        1: pattern1,
        2: pattern2,
        3: pattern3,
        4: pattern4
             }
    tests_results = [f'{re.search(tests[i], password)}' for i in range(len(tests))]
    c = True if 'None' not in tests_results[1:] else False
    test_zero = tests_results[0] == 'None'
    ans = test_zero and c
    return ans


while True:
  psswrd = password_generation()
  if check_password(psswrd):
    print(psswrd)
    break
````

Предположим, что у нас есть директория с большим количеством текстовых файлов -- N. Мы хотели бы один раз по ним пройтись и научиться выдавать документы, в которых есть то или иное ключевое слово.

Алгоритм построения инвертированного индекса:

* Построить словарь соответствия, где каждому уникальному слову будет поставлен в соответствие некоторый внутренний id: {word_1 : 0, word_2: 1 ...}. Можно делать "на лету"
* Прочитать все документы и собрать все пары (wordID, docID). Для простоты предполагаем, что количество слов в документах нам не важно, важен сам факт, что в документе docID есть слово wordID (hence, какой тип данных нам будет удобен?)
* Сгруппировать пары так, чтобы у нас получились новые пары (wordID, [list of docIDs])

In a nutshell, каждому документу в базе сопоставляется вектор, и тогда поиск похожих документов будет равносилен вычислению некоторой метрики расстояния между векторами документов и взятия ближаших. Часто используемая метрика -- косинусное расстояние. Один из простых способов задания вектора:

* взять топ N самых популярных слов в базе, где N определит нашу размерность данных
* сопоставить каждому слову слову позицию от 0 до N
* сформировать вектор как `[tf-idf(word_0), ..., tf-idf(word_N)]`

Понятно, что у данного способа много недостатков, но на практике это может быть достаточно неплохим бейзлайном. Если вам интересно, можете реализовать такую логику в рамках класса DirectoryIndex вместо второго бонуса (только реализовать самим и аккуратно!)

Ниже представлена реализация данного класса (с рекурсивным обходом всех папок, удалениеем стоп-слов из всех найденных файлов и попыткой угадать их кодировку, считая, что она одна и та же для всех текстовых файлов в переданном пути), а также добавлены аттрибуты для расчета tf-idf слова документа и косинусной метрики как двух файлов, так и для каждого файла из всех.

````python
class DirectoryIndex:
    def __init__(self, dirpath: str, mode=False, encoding: str='utf-8'):
        self.__d = defaultdict(set)
        self.stop_words = set(stopwords.words('english'))
        self.__list_of_files_txt = []

        if mode:
            self.find_all_txt_files_in_path(dirpath)
        else:
            for file_name in os.listdir(dirpath):
                self.__list_of_files_txt.append(f'{dirpath}/{file_name}')

        self.encoding = f"cp12{''.join(self.guess_encoding())}"
        
        for i in range(len(self.__list_of_files_txt)):
            file_name = self.__list_of_files_txt[i][0]
            self.for_one_file(file_name, int(file_name.split('/')[-1].split('.')[0]), i)


    def guess_encoding(self):
      file_path = self.__list_of_files_txt[0][0]
      for i in range(10):
          for j in range(10):
            try:
                f = open(file_path, 'r', encoding=f'cp12{i}{j}')
                return f'{i}', f'{j}'
            except LookupError as e:
                pass


    def delete_stop_words_from_file(self, text: list) -> list:
        filtered_words = []
        for line in text: 
          words = set(self.split_into_words(line))
          filtered_words.append([word for word in words if word not in self.stop_words])
        return filtered_words


    @staticmethod
    def split_into_words(line: str):
        word_regex_improved = r"(\w[\w']*\w|\w)"
        word_matcher = re.compile(word_regex_improved)
        return word_matcher.findall(line)


    @private
    def for_one_file(self, file_path: str, file_id: int, k: int):
        # we can substitute the file with it copy without stop-words
        f = open(file_path, 'r', encoding=self.encoding)
        text = self.delete_stop_words_from_file(f.readlines())

        words_count = 0
        for line in text:
            words_count += len(line)
            for word in line:
                self.__d[word.lower()].add(file_id)
        # add number of words of certain file together with its path
        self.__list_of_files_txt[k].append(words_count)


    @private
    def find_all_txt_files_in_path(self, path: str):
        for i in os.listdir(path):
            if os.path.isdir(path + '\\' + i):
                self.find_all_txt_files_in_path(path)
            elif os.path.splitext(os.path.join(path, i))[1] == '.txt':
                self.__list_of_files_txt.append([os.path.join(path, i)])



    def find_documents(self, word: str) -> list:
        return sorted(list(self.__d[word]))
    
    def update(self, filepath: str, encoding: str='utf-8'):
        self.__list_of_files_txt.append([filepath])
        f = open(filepath, 'r', encoding=encoding)
        text = self.delete_stop_words_from_file(f.readlines())

        words_count = 0
        for line in text:
            # words = set(self.split_into_words(line))
            words_count += len(line)
        self.__list_of_files_txt[-1].append(words_count)

    
    # #@property
    # def inverted_index(self):
        # pass

    
    def slice_of_words(self, stop=10, start=0) -> list:
        return sorted(self.__d.keys(), key=lambda x: len(self.__d[x]), reverse=True)[start: stop + 1]


    def tf(self, my_word: str, document_path: str) -> float:
      # we can substitute the file with it copy without stop-words
        f = open(document_path, 'r', encoding=self.encoding)
        text = f.readlines()

        word_count = 0
        my_word_count = 0
        for line in text:
            words = set(self.split_into_words(line))
            word_count += len(words)
            for word in words:
                if word == my_word:
                     my_word_count += 1
        return my_word_count / word_count


    def idf(self, my_word: str) -> float:
        return log(len(self.__list_of_files_txt) / len(self.find_documents(my_word)))

    
    def tf_idf(self, my_word: str, document_path: str) -> float:
        return self.tf(my_word, document_path) * self.idf(my_word)


    def vectorise_one_file(self, file_path: str, words: list) -> list:
        v = []
        for word in words:
            v.append(self.tf_idf(word, file_path))
        return v


    def vactorize_all_files(self, n: tuple, list_of_files=None) -> dict:
        if list_of_files == None:
            list_of_files = self.__list_of_files_txt
        d = defaultdict(list)
        words = self.slice_of_words(n[1], n[0])
        for file_name in list_of_files:
            d[file_name[0]] = np.array(self.vectorise_one_file(file_name[0], words))
        return d


    def cosine_metrics_for_two_docs(self, v1: np.array, v2: np.array) -> int:
        return v1 @ v2 / (LA.norm(v1) * LA.norm(v2))
    

    def count_cosine_metrics_for_each_doc(self, n: tuple):
        d = defaultdict(list)
        all_file_vectors = self.vactorize_all_files(n)
        for file_name in self.__list_of_files_txt:
            v1 = all_file_vectors[file_name[0]]
            d[file_name[0]] = [self.cosine_metrics_for_two_docs(v1, all_file_vectors[other_file[0]]) for other_file in self.__list_of_files_txt if file_name[0] != other_file[0]]
        return d

      


# могут быть проблемы с приватностью некоторых аттрибутов, которые я не учитывал
# (но это монжно исправить, используя @privat или _attribute, ___attribute)
# также в процессе создания экземпляра класса была произведена преобработка каждого файла: из него были удалены все стоп-слова,
# потому что иначе было слишком много слов, для которых tf-idf = 0 
````
