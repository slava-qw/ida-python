```python
N = 2
A = [0, 1, 2, 3, 4, 5]



def search(n, a):
  """
  Find number n in array a using binary search with recursion (O(log(len(a))))
  """
  
  if len(a) == 1:
      if n == a[0]:
          return a[0]
      else:
        return
  if n <= a[len(a)//2 - 1]:
    a = a[:len(a)//2]
  else:
    a = a[len(a)//2:]
  return search(n, a)

# some problems with N, when N - A[i] = N / 2
# if N / 2 % 10 == 0:
# else:


# Find number N - A[i], if such number exists, add it to the l together with the A[i]
l = []
for i in range(len(A)):
    num = N - A[i]
    if num >= N//2:
      scn_num = search(num, list(A[:i] + A[i + 1:]))
      if scn_num != None:
        l.append((A[i], scn_num))
    else:
        break

if bool(l):
  print(f'There are some numbers that fit the condition: {l}')
else:
  print('There are no such numbers')
```

## 3.3

**Три числа с суммой $N$**

Пусть дан отсортированный список чисел `[10, 10, 15...]` и некоторое число $N$. Проверить, есть ли в нем $3$ числа, которые в сумме дают $N$. Вывести такие "*триплеты*" чисел.

```python
def search2_pt(n, a):
  """
  Find number n in array a using the two-pointer technique (O(len(a)))
  """
  l = []
  i = 0
  j = len(a) - 1
  while len(a[i:j+1]) != 1:
    if a[i] + a[j] > n:
      j -= 1
    elif a[i] + a[j] < n:
      i += 1 
    else:
      l.append((a[i], a[j]))
      if len(a[i:j + 1]) == 2:
          break
      else:
          i += 1
          j -= 1
  return l


N = 4
A = [0, 0, 0, 1, 1, 1, 2, 2, 3, 4, 5, 7, 8, 9, 10]
s = []

for i in range(len(A)):
    num = N - A[i]
    B = A[:i] + A[i + 1:]
    if num > N//2:
        s.append((A[i], search2_pt(num, B)))
    else:
        break

# make the found answer looks better
p = []
for i in range(len(s)):
  for j in range(len(s[i][1])):
    p.append((s[i][0], s[i][1][j][0], s[i][1][j][1]))
if bool(p):
  print(f'There are some triplets that fit the condition: {set(p)}') # проблема удаления одинаковых триплетов, которые есть в списке
else:
  print('There are no such triplets')
```

```python
# код для уникальных триплетов
# (такой же как и выше только для поиска двух чисел используется линейный алгоритм и в конце удаляются повторяющиеся элементы)
def search2(n, a):
  """
  Find all pairs of numbers in array a, which sum is equal n using the modified two-pointer technique (O(len(a)))
  """
  l = []
  i = 0
  j = len(a) - 1
  while len(a[i:j+1]) != 1:
    if a[i] + a[j] > n:
      j -= 1
    elif a[i] + a[j] < n:
      i += 1 
    else:
      l.append((a[i], a[j]))
      if len(a[i:j + 1]) == 2:
          break
      else:
          i += 1
          j -= 1
  return l


N = 4
A = [0, 0, 1, 1, 1, 2, 2, 3, 4, 5, 7, 8, 9, 10]
s = []

# Find number num = N - A[i], which can be present as sum of two numbers from array A, if so, add them to the s together with the A[i]
for i in range(len(A)):
    num = N - A[i]
    B = A[:i] + A[i + 1:]
    if num > N//2:
        s.append((A[i], search2(num, B)))
    else:
        break

if bool(s):
  # make the found answer looks better
  p = []
  for i in range(len(s)):
    for j in range(len(s[i][1])):
      p.append((s[i][0], s[i][1][j][0], s[i][1][j][1]))

  # find the unique triplets
  t = sorted(list(map(sorted, set(p))))
  i = 0
  while i != len(t) - 1:
    if t[i] == t[i + 1]:
      del t[i]
    else:
      i += 1
  print(f'There are some unique triplets that fit the condition: {t}')
else:
  print('There are no such triplets')
```

```python
from math import log
def preprocess_text(text: str):
  word = r"(\w[\w']*\w|\w)" # also fits for words with 's
  return re.findall(word, text)


def preprocess_corpus(corpus: str):
  '''
  Split corpus of text to sentences.
  '''
  return re.split(r'(?<=\w[:;.!?]) ', corpus) 
  # проблемы с разделением вот такого предложения:
  # 'Как выжить в современной России. У нас есть три способа: 1) стать жертвой маньяка; 2) стать жертвой другого маньяка, которого мы убьем; 3) стать жертвой не маньяка.',
  

def calc_tf(term: str, text: str):
  '''
  Claculate tf for term in the sentence.
  '''
  sentences = preprocess_corpus(text)
  i = 0
  sentence = ''
  for s in sentences:
    if s.find(term) != -1:
      sentence = sentences[i]
      break
  words = preprocess_text(sentence)
  return 0.5 + 0.5 * (words.count(term) / len(words))


def calc_idfs(corpus):
  whole_text = ' '.join(corpus)
  A = preprocess_corpus(whole_text)
  N = len(A)
  k = 0
  d = defaultdict(int)
  
  for sent in A:
    for word in preprocess_text(sent):
      if word.isalpha():
        d[word.lower()] += 1

  for word in d:
    c = 0
    for sent in A:
      if word in sent:
        c += 1

    d[word] = log(N / (1 + c))

  return d

def calc_tfidf(term: str, text: str, precomputed_idfs: dict):
    return calc_tf(term, text) * precomputed_idfs[term]
```

```python
from functools import lru_cache
import time
```

```python
def clock(active=True, unit='s'):

    def decorate(func):
        
        if unit == 's':
            modifier = 1
        elif unit == 'm':
            modifier = 60
        elif unit == 'ms':
            modifier = 0.001
        elif unit == 'h':
            modifier = 3600
        
        
        @wraps(func)
        def clocked(*args, **kwargs):
            if not active:
                return func(*args, **kwargs)
            
            t_start = time.time()
            res = func(*args, **kwargs)
            total_time = time.time() - t_start

            print(f'{func.__name__}({args}) -> {res} executed in {total_time / modifier:.3f} {unit}')

            return res

        return clocked
    
    return decorate
```

```python
from functools import wraps

def my_cache_deco(N=None, K=None, cache_type=1):

  def decororate(func):
    cache = {}
    root = []
    root[:] = [root, root, None, None]
    PREV, NEXT, KEY, RESULT = 0, 1, 2, 3
    full = False


    @wraps(func)
    def mem(*args):
      if N == None: # no cache
          result = func(*args)
          return result

      elif N == -1:  # cache without limits
          key = args[0]
          result = cache.get(key, None)
          if result != None:
              return result
          
          result = func(*args)
          cache[key] = result
          return result

      
      if cache_type == 1:
          nonlocal root, full
          key = args[0]
          link = cache.get(key, None)

          if link is not None:
              link_prev, link_next, _key, result = link
              link_prev[NEXT] = link_next
              link_next[PREV] = link_prev
              last = root[PREV]
              last[NEXT] = root[PREV] = link
              link[PREV] = last
              link[NEXT] = root
              return result
          
          result = func(*args)

          if key in cache:
              pass
          elif full:
              oldroot = root
              oldroot[KEY] = key
              oldroot[RESULT] = result
              root = oldroot[NEXT]
              oldkey = root[KEY]
              oldresult = root[RESULT]
              root[KEY] = root[RESULT] = None
              del cache[oldkey]
              cache[key] = oldroot
          else:
              last = root[PREV]
              link = [last, root, key, result]
              last[NEXT] = root[PREV] = cache[key] = link
              full = (len(cache) >= N)
          return result

      if cache_type == 2:
          pass
        

    return mem
  
  return decororate
```

```python
@clock()
@my_cache_deco(N=4, cache_type=1)
def strange_factorial(n):
  time.sleep(1)
  if n == 0:
    return 1
  elif n == 1:
    return 1
  elif n == 2:
    return 2
  
  return strange_factorial(n - 1) + strange_factorial(n - 1)
```

Сгенерируйте случайный пароль длины N. Убедитесь, что получившийся пароль на самом деле хороший. В нашем случае это:

* Содержит хотя бы две заглавные буквы подряд
* Содержит хотя бы две строчные буквы подряд
* Не содержит пробельные спецсимволы (\t, \n и подобные)
* Не содержит трех цифр подряд
* Содержит и русские, и английские буквы

Ваш код должен генерировать пароли, пока они не удовлетворят заданным условиям.

hint: используйте модуль random

```python
import string
import random
import re


cyrillic_lower = [(lambda c: chr(c))(i) for i in range(1072, 1104)]
cyrillic_upper = [(lambda c: chr(c))(i) for i in range(1040, 1072)]
cyrillic_ansi = cyrillic_lower + cyrillic_upper
cyrillic_ansi_str = ''.join(cyrillic_ansi)

allowed_symbols = string.ascii_letters + string.digits + cyrillic_ansi_str


def password_generation(N=10):
    password = "".join([random.choice(allowed_symbols) for n in range(N)])
    return password


def check_password(password):
    pattern1 = re.compile('([A-Z]{2,}|[А-Я]{2,})')  # test1: True
    pattern2 = re.compile('([a-z]|[а-я]){2,}')  # test2: True
    pattern_zero = re.compile('\d{3,}')  # false
    pattern3 = re.compile('[A-Za-z]')  # true
    pattern4 = re.compile('[А-Яа-я]')  # true
    tests = {
        0: pattern_zero,
        1: pattern1,
        2: pattern2,
        3: pattern3,
        4: pattern4
             }
    tests_results = [f'{re.search(tests[i], password)}' for i in range(len(tests))]
    c = True if 'None' not in tests_results[1:] else False
    test_zero = tests_results[0] == 'None'
    ans = test_zero and c
    return ans


while True:
  psswrd = password_generation()
  if check_password(psswrd):
    print(psswrd)
    break
```
